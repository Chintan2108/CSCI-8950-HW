### Part 1: Instantenous Field of View (IFOV) Calculation

#### Q1: Calculate IFOV
* α = Optic: 25 degree
* h = Height above target: 60 cm
* d = (IFOV) Instantaneous Field of View: = ______________ cm?

d = 2 * h * tan(α/2)

  = 2 * 60 cm * tan(12.5deg)
  
  = **26.60 cm**

#### Q2: Calculate IFOV
* α = Optic: 25 degree
* h = Height above target: 20 m
* d = (IFOV) Instantaneous Field of View: = ______________ cm?

d = 2 * h * tan(α/2)

  = 2 * 20 m * tan(12.5deg)
  
  = **8.88 m**

#### Q3: Calculate *h* (height of capture)
* α = Optic: 25 degree
* h = Height above target: ______________ m?
* d = (IFOV) Instantaneous Field of View: = 30 m

d = 2 * h * tan(α/2)

⇒ h = (2 * tan(α/2))

   = 30 m / (2 * tan(12.5deg)) 

   = **67.57 m** <br><br>

#### Q2:  Calculate NDVI, VARI, WDRVI for both soybean dryland and soybean irrigated (10 points). Wavelength range to be used for calculating these indices are given below: 

Green, Red and NIR bands were derived by averaging the said wavelengths as instructed. Based on these, the following vegetation indices were calculated:
1. NDVI = (NIR - Red)/(NIR + Red)
2. WDRVI = (α x NIR - Red)/(α x NIR + Red)
3. VARI = (Green - Red)/(Green + Red)

WDRVI was calculated at α=0.2. In case of VARI, typically blue reflectance is also subtracted to account for the atmospheric effects and make the index "atmopspherically invariant", but this is not a healthy practice to get rid of the atmopsheric effects (as discussed in class). So only the Green and Red bands have been used to compute VARI. #### Q7: Discuss which vegetation index works well to predict GLAI and why? (10 points).

* NDVI and WDRVI(α=0.2) were used to calibrate a GLAI model by fitting these indices for irrigated soyabean plantation GLAI values, which yielded models with correlation coefficients of 93.52% and 94.04% respectively. 


* Validation of the NDVI and WDRVI based GLAI models on an unseen dataset (soyabean dryland plantations) revealed confidence scores of 95.61% and 94.32% respectively. 


* Both models have shortcomings in the form of general overestimation of GLAI values (Figure 9). Although the validation confidence score of NDVI model is more than that of WDRVI model, I would say that ***WDRVI model works better to predict GLAI***. This is mainly because of the it's higher calibration coefficient (R<sup>2</sup>) of 94.04%. This essentially means that WDRVI-based model is slightly more capable of quantifying GLAI overall as compared to the NDVI-based model (to be specific, 0.52% better). Moreover, a simple trend analysis of the GLAI actual values and predicted values from both models shows that overall, the trendline of the WDRVI-based predictions is closer to the actual values as compared to the NDVI-based predictions (Figure 9).
